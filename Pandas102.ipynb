### ต่อแกน X

```python
personal_data = data_df[['No.' , 'sex' , 'age' , 'nationality']]
personal_data.head()
```

```python
covid_data = data_df[['No.' , 'announce_date' , 'province_of_isolation' , 'province_of_onset' , 'risk']].sample(frac=1) ##sample(frac=1) คิอการสุ่มสลับ row
covid_data.head()
```

###เอาตารางสองอันมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data)
personalcovid_data.head()
```

```python
personalcovid_data = covid_data.merge(personal_data)
personalcovid_data.head()
```

### Save Table

```python
personalcovid_data.to_csv(os.path.join(path, 'my_processed_data.csv') ) ## save ตารางเอาไปใช้ที่อื่น
```

##List(row oriente)

```python
data = [['Jones LLC',150,200,140],
        ['Alpha Co',200,210,215],
        ['Blue Inc',50,90,95]]
header = ['account','Jan','Feb','Mar']
```
