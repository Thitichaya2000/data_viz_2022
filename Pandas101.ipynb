###Quiz (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)

1. คำนวณ อายุเฉลี่ย ของผู้ชาย ในข้อมูลทั้งหมด
2. คำนวณ อายุเฉลี่ย ของผู้ชาย ในจังหวัดขอนแก่น
3. หาจำนวนผู้ป่วยที่เป็นคน ที่ไม่ใช่คนไทย

```python
this_data = data_covid[['sex','age','province_of_onset']] ##เลือกเฉพาะ column ที่ต้องการ โดยเลือก sex , age และ province_of_onset
```

```python
this_data #แสดงผล
```

```python
##1.คำนวณ อายุเฉลี่ย ของผู้ชาย ในข้อมูลทั้งหมด
male = this_data[this_data['sex']=='ชาย'] ##เลือกเฉพาะที่เป็นผู้ชาย สนใจแต่ sex ใส่ใน male 
male['age'] 
```

```python
sum = 0 #คำนวณอายุเฉลี่ย มีตัวแปรมารับ โดย sum=0
N = 0 #มีกี่คน
for a in male['age']: #ทำการวนลูป
    
    if a > 0: #ถ้า a > 0 เป็นอายุคนปกติ
        sum += a # sum = sum + a #มารวมใน sum
        N += 1 #คำนวณจำนวน +1 เพิ่มเรื่อยๆ

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย {sum/N}')
```

```python
##2.คำนวณ อายุเฉลี่ย ของผู้ชาย ในจังหวัดขอนแก่น
male_KK = male[male['province_of_onset']=='ขอนแก่น'] #สนใจแต่จังหวัดขอนแก่น

sum = 0
N = 0
for a in male_KK['age']: #ทำการวนลูป
    
    if a > 0:
        sum += a # sum = sum + a
        N += 1 #คำนวณจำนวน +1 เพิ่มเรื่อยๆ

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย ในจังหวัดขอนแก่น {sum/N}')
```

```python
##3.หาจำนวนผู้ป่วยที่เป็นคน ที่ไม่ใช่คนไทย
data_outlander = data_covid[data_covid['nationality'] != 'Thailand']
data_outlander
```

###Quiz ตัดตารางออกมาเฉพาะปี 2021 announce_date ในปี 2021

Hint

* วนลูปหา index ของปี 2021
* ตัดตารางมาเฉพาะ ปี 2021

```python
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
```

##HW กลุ่ม 2 || เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20

```python
data_covid #แสดงข้อมูล data_covid
```

```python
HW = list()  #ให้ HW เป็น list ว่าง
for each_row in data_covid.iterrows():  #วนลูปในตาราง data_covid 
    if (int(each_row[1]['announce_date'].split('/')[1] ) >=1) and (int(each_row[1]['announce_date'].split('/')[1] ) <=20):  #สร้างเงื่อนไขให้ announce_date มากกว่าหรือเท่ากับ 1 และ น้อยกว่าหรือเท่ากับ 20
        HW.append(True)  #หากเงื่อนไข  = true ให้เพิ่มลงใน HW
    else:
        HW.append(False)   #หากไม่ตรงเงื่อนไข  = false  ไม่ให้เพิ่มลงใน HW

data_covid[HW]
```
