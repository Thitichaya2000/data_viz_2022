## ต่อตาราง


```python
covid = pd.concat([data_covid2020,data_covid2021_1,data_covid2021_2,data_covid2022]) ## ใช้คำสั่ง pd.concat ในการต่อตารางทั้ง 4 ตารางได้แก่ data_covid2020,data_covid2021_1,data_covid2021_2 และ data_covid2022 ## โดยตั้งชื่อตารางที่ต่อว่า covid
covid ## แสดงตาราง covid
```

## เลือกเฉพาะคอลัมน์ที่สนใจ

```python
covid_data = covid[['announce_date','sex','age','nationality','risk','province_of_onset']] ## เลือกเฉพาะคอลัมน์ที่สนใจในตาราง covid คือ announce_date','sex','age','nationality','risk' และ 'province_of_onset ##โดยตั้งชื่อตารางนี้ว่า covid_data
covid_data ## แสดงตาราง covid_data
```

## ตรวจสอบค่า Missing

```python
covid_data.isnull().any() ## ดูว่าแต่ละ column มีค่า missing มั้ย
## จะเห็นได้ว่าคอลัมนื announce_date เป็นคอลัมน์เดียวที่ไม่มีค่า missing
```

```python
## ทำการ clean ข้อมูลโดยการแทนค่า NaN เป็น "ไม่ระบุ" ในคอลัมน์ sex  nationality  risk และ  province_of_onset  
covid_data[['sex']] = covid_data[['sex']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
covid_data[['nationality']] = covid_data[['nationality']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
covid_data[['province_of_onset']] = covid_data[['province_of_onset']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
covid_data[['risk']] = covid_data[['risk']].fillna('ไม่ระบุ') ##แทนค่า NaN เป็น ไม่ระบุ
```
